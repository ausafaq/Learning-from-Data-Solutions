```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate a data set of 20. 
#for simplicity, 10 in the first quadrant, another 10 in the third quadrant 
X1 = []
Y1 = []
X2 = []
Y2 = []

#weight = [0,0,0]

for i in range(10):
    X1.append(random.uniform(0,1))
    Y1.append(random.uniform(0,1))
    X2.append(random.uniform(-1,0))
    Y2.append(random.uniform(-1,0))


#label the data
data1 = [np.array([1,X1[i],Y1[i],1]) for i in range(10)]
data2 = [np.array([1,X2[i],Y2[i],-1]) for i in range(10)]
data = data1 + data2
plt.plot(X1,Y1,'ro')
plt.plot(X2,Y2,'go')

# y = -x 
x = np.linspace(-1,1,20)
y = (-1 * x)
plt.plot(x, y, color='purple')
plt.show()
```

```python
def pla_plot(d_list,x_color,y_color):
    x_point = []
    y_point = []
    weight = [0,0,0]
    count = 0
    for i in range(len(data)):
        point = data[i]
        point.tolist()
        if(np.sign(np.matmul(point[0:3],weight)) != point[3]):
            weight = np.add(weight,(np.dot(point[3],point[0:3])))
            axes = plt.gca()
            plt.plot(X1,Y1,x_color)
            plt.plot(X2,Y2,y_color)
            plt.plot(x, y, color='purple')
            x_h = np.array(axes.get_xlim())
            y_h = (-(weight[1]/weight[2])*x_h)-(weight[0]/weight[2])
            plt.plot(x_h,y_h,color='black')
            plt.show()
            count+=1
    print("Total Number of Iterations: ", +count)       
pla_plot(data,'ro','go')

        
```
